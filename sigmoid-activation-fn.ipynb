```python
import numpy as np
```

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Generate values
z = np.linspace(-10, 10, 100)
sigmoid_values = sigmoid(z)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(z, sigmoid_values)
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("σ(z)")
plt.grid(True)
plt.show()
```
